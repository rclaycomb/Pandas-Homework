```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
%matplotlib inline

# Setting file Paths
whale_returns = Path("whale_returns.csv")
# Converting to DateTimeIndex
whale_returns = pd.read_csv(whale_returns, index_col="Date", infer_datetime_format=True, parse_dates=True)
# Detecting null valves and removing
whale_returns.isnull().mean() 
whale_returns.isnull().sum()
whale_returns = whale_returns.dropna()



# Setting file path
algo_returns = Path("algo_returns.csv")
# Converting to DateTimeIndex
algo_returns = pd.read_csv(algo_returns, index_col="Date", infer_datetime_format=True, parse_dates=True)
# Detecting null valves and removing
algo_returns.isnull().mean() 
algo_returns.isnull().sum()
algo_returns = algo_returns.dropna()



# Setting file path
sp500_history = Path("sp500_history.csv")
# Converting to DateTimeIndex
sp500_history = pd.read_csv(sp500_history, index_col="Date", infer_datetime_format=True, parse_dates=True)
# Removed dollar signs from the Close column in sp500_history
sp500_history['Close'] = sp500_history['Close'].str.replace('$', '').astype(float)
# Converting sp500 closing prices to daily returns
sp500_history_dr = sp500_history.pct_change()
# Detecting null valves and removing
sp500_history_dr.isnull().mean() 
sp500_history_dr.isnull().sum()
sp500_history_dr= sp500_history_dr.dropna()
# Changing column name for sp500 from closing to S&P 500
sp500_history_dr.columns = ["S&P 500"]

# Combining all files
combined_df = pd.concat([whale_returns, algo_returns, sp500_history_dr], axis="columns", join="inner")
combined_df.head()
```

```python
# Calculate and ploting cumulative returns
cumulative_returns = (1 + combined_df).cumprod()
cumulative_returns.head()
```

```python
# Plotting cumulative returns
cumulative_returns.plot(figsize=(20,10))
```

* Does any portfolio outperform the S&P 500?

In late 2015 the S&P 500 did outperform every portfolio, but toward the end of the year and into 2016 things did change. The Algo 1 and very briefly Berkshire Hathaway portfolio passed the S&P 500. Once we got to end of 2016 and into 2017 every portfolio surpassed the S&P 500 besides the Paulson & CO which remained very close to the S&P 500 for the remainder of the time.

---

```python
# Creating box plot to determine which portfolios has the greatest and smallest spread
cumulative_returns.plot.box(figsize=(20,10))
```

* Which box has the largest spread? Which has the smallest spread?

Berkshire Hathaway INC has the largest spread. Soros Fund Management LLC has the smallest spread.

---

```python
#Calculating the standard deviation for each portfolio
returns_std = combined_df.std()
returns_std
```

* Which portfolios are risker than the S&P 500?

Only the Berkshire Hathaway INC and Tiger Global Management LLC portfolios are risker than the S&P 500.

---

```python
# Plotting rolling standard deviation of all portfolios in 180 day window
combined_df.rolling(window=180).std().plot(figsize=(20,10))
```

* Does the risk increase for each of the portfolios at the same time risk increases in the S&P?

Yes they do for the most part. They all have mostly the same pattern except for the Tiger Global Management LLC portfolio which in roughly the first quarter of 2017, the risk went up quite a bit when all the others were going down. Also at the end of 2017 it had a major drop off as the others were more flat lining.

The Soros fund management LLC and Algo 2 are the closest that mimic the S&P 500.

Yes it does. Algo 1 does have a few spots where it does spike alot higher than the S&P 500 but does for the most part follow the S&P 500.

```python
# Calculating sharpe ratios
sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Using bar plot to plot sharpe ratios
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

* Determine whether the algorithmic strategies outperform both the S&P 500 and the whales portfolio.

The algorithmic strategies do outperform the S&P 500 and most of the whale portfolio besides the Berkshire Hathaway INC which only surpasses Algo 2 by .12.

```python
# Adding in new files 
boeing_data = Path("Boeing_data.csv")
boeing_data = pd.read_csv(boeing_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
boeing_data_dr = boeing_data.pct_change()
boeing_data_dr.isnull().mean() 
boeing_data_dr.isnull().sum()
boeing_data_dr = boeing_data_dr.dropna()
boeing_data_dr.columns = ["Boeing"]

coca_cola_data = Path("Coca_cola_data.csv")
coca_cola_data = pd.read_csv(coca_cola_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
coca_cola_data_dr = coca_cola_data.pct_change()
coca_cola_data_dr.isnull().mean() 
coca_cola_data_dr.isnull().sum()
coca_cola_data_dr = coca_cola_data_dr.dropna()

netflix_data = Path("Netflix_data.csv")
netflix_data = pd.read_csv(netflix_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
netflix_data_dr = netflix_data.pct_change()
netflix_data_dr.isnull().mean() 
netflix_data_dr.isnull().sum()
netflix_data_dr = netflix_data_dr.dropna()

new_combined_df = pd.concat([boeing_data_dr, coca_cola_data_dr, netflix_data_dr], axis="columns", join="inner")
new_combined_df.head()
```

```python
# Combining first set of data with new set of data
combined_portfolios = pd.concat([combined_df, new_combined_df], axis="columns", join="inner")
combined_portfolios.head()
```

```python
# Calculate and ploting cumulative returns with new data
new_cumulative_returns = (1 + combined_portfolios).cumprod()
new_cumulative_returns.head()
```

```python
# Plotting cumulative returns with new data
new_cumulative_returns.plot(figsize=(20,10))
```

```python
# Creating box plot to determine which portfolios has the greatest and smallest spread with new data
new_cumulative_returns.plot.box(figsize=(20,10))
```

```python
#Calculating the standard deviation for each portfolio with the new data
new_returns_std = combined_portfolios.std()
new_returns_std
```

```python
# Plotting rolling standard deviation of all portfolios in 180 day window with the new data included
combined_portfolios.rolling(window=180).std().plot(figsize=(20,10))
```

```python
# Creating correlation table with new added data
new_correlation = combined_portfolios.corr()
new_correlation
```

```python
# Calculating sharpe ratios with new added data
new_sharpe_ratios = (combined_portfolios.mean() * 252) / (combined_portfolios.std() * np.sqrt(252))
new_sharpe_ratios
```

```python
# Using bar plot to plot sharpe ratios with new added data
new_sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```
